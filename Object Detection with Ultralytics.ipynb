Import necessary libraries (especially Ultralytics)

```python
from ultralytics import YOLO
import os
```

Download YOLOv11 model (done through ultralytics library)

```python
model = YOLO("yolo11n.pt")
```

Detect objects with the model using your own images

```python
def predict_image(model, img_list):
  results = model.predict(img_list)
  if results:
    for result in results:
      result.show()
```

Add any image of your choice to a folder and add the path in the cell below

```python
imgs_dir = "/content/drive/MyDrive/AI Club Projects/Object Detection with Ultralytics/imgs" #Insert your own path directory here
```

```python
imgs_list = [imgs_dir + "/" + img for img in os.listdir(imgs_dir)]
```

Object Detection Results Below..

```python
predict_image(model, imgs_list)
```
